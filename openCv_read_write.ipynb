```python
#edge detection
blurred = cv2.GaussianBlur(gray_bike2_read,(3,3),0)
```

```python
canny = cv2.Canny(blurred,threshold1=180, threshold2=200)
```

```python
resized = cv2.resize(canny,(int(2190/2),int(2738/2)))
cv2.imshow(resized)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
